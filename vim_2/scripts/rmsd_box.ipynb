```python
import numpy as np
import MDAnalysis as mda
import MDAnalysis.analysis.rms
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
rmsd_lig_file = "../nvt/rmsd_lig.xvg"
lig_reader = mda.auxiliary.XVG.XVGReader(rmsd_lig_file)
rmsd_file = "../nvt/rmsd.xvg"
reader = mda.auxiliary.XVG.XVGReader(rmsd_file)

rmsd_lig = [step.data[1] for step in lig_reader]
rmsd = [step.data[1] for step in reader]

rmsd_dictionary = {"rmsd": [rmsd, rmsd_lig]}

fig = plt.figure(figsize=(10, 10))
sns.set(context="notebook", palette="colorblind", style="ticks", font_scale=2)

median_line_properties = dict(linestyle='-', linewidth=2.5, color="#D0006F")
xtick_positions = np.arange(1, 3, 1)
plt.boxplot(rmsd_dictionary["rmsd"], medianprops=median_line_properties)
plt.xticks(ticks=xtick_positions, labels=[r"C$_\alpha$", "Ligands"])
# plt.xlabel("")
plt.ylabel(f"RMSD ($\AA$)")
# plt.savefig(f"../../plots/rmsd/{stage}/box_plot_rmsd_{stage}.pdf")
```
