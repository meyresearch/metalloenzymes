```python
system_1 = bss.IO.readMolecules([dft_1_path+"vim2_dry.prmtop", dft_1_path+"vim2_dry.inpcrd"])
system_2 = bss.IO.readMolecules([dft_2_path+"vim2_dry.prmtop", dft_2_path+"vim2_dry.inpcrd"])
```

```python
system_1.getMolecule(0)
```

```python

system_box_min, system_box_max = system_1.getAxisAlignedBoundingBox()
system_box_size = [y - x for x, y in zip(system_box_min, system_box_max)]
system_box_sizes = [x + 20 * bss.Units.Length.angstrom for x in system_box_size]

box, angles = bss.Box.cubic(max(system_box_sizes))
print("solvating bound system")
system_1_solvated = bss.Solvent.solvate("tip3p", molecule=system_1, box=box, angles=angles)
```

```python

system_box_min, system_box_max = system_2.getAxisAlignedBoundingBox()
system_box_size = [y - x for x, y in zip(system_box_min, system_box_max)]
system_box_sizes = [x + 20 * bss.Units.Length.angstrom for x in system_box_size]

box, angles = bss.Box.cubic(max(system_box_sizes))
print("solvating bound system")
system_2_solvated = bss.Solvent.solvate("tip3p", molecule=system_2, box=box, angles=angles)
```

```python
sys_1_path = "../vim_2/inputs/protein/"
sys_2_path = "../vim_2/inputs/protein/"
save_1 = sys_1_path + "system_1_solvated"
save_2 = sys_2_path + "system_2_solvated"
bss.IO.saveMolecules(save_1, system_1_solvated, ["PRM7", "RST7"])
bss.IO.saveMolecules(save_2, system_2_solvated, ["PRM7", "RST7"])
bss.IO.saveMolecules(save_1, system_1_solvated, ["GroTop", "Gro87"])
bss.IO.saveMolecules(save_2, system_2_solvated, ["GroTop", "Gro87"])
```
