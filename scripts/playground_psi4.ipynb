```python
import psi4
```

```python
h2o = psi4.geometry("""
O
H 1 0.96
H 1 0.96 2 104.5
""")
```

```python
psi4.set_memory("500 GB")
psi4.set_num_threads(30)
```

```python
psi4.optimize(name="b3lyp/6-31G*", molecule=h2o)
```

```python
file = "/home/jguven/projects/parameterisation_with_psi4/vim2_small_opt.xyz"
with open(file, "r") as f:
    xyz_contents = f.read()
# define geometry from a file
geometry = psi4.core.Molecule.from_string(xyz_contents, dtype="xyz")
print(geometry.molecular_charge()) # get charge
print(geometry.multiplicity()) # get spin multiplicity
```

```python
psi4.set_memory("12 GB")
psi4.set_num_threads(30)
psi4.optimize(name="B3LYP/6-31G*", molecule=geometry)
```
