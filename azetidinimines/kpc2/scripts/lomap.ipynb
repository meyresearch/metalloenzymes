```python
path_to_ligands = "../vim_2/inputs/ligands/"
working_directory = "/home/jguven/projects/metalloenzymes/vim_2/inputs/ligands/"
```

```python
ligand_files = sorted(glob.glob(f"{path_to_ligands}/docked_*.sdf"))
print(ligand_files)
ligands = []
ligand_names = []

for filepath in ligand_files:
    ligands.append(bss.IO.readMolecules(filepath)[0])
    ligand_names.append(filepath.split("/")[-1].replace(".sdf",""))

transformations, lomap_scores = bss.Align.generateNetwork(ligands, plot_network=True, names=ligand_names)
```

```python
for i in range(len(ligands)):
    print(i+1, ligand_files[i])
```

```python
pert_network_dict = {}
transformations_named = [(ligand_names[transf[0]], ligand_names[transf[1]]) for transf in transformations]
for transf, score in zip(transformations_named, lomap_scores):
    print(transf, score)
    pert_network_dict[transf] = score
```
